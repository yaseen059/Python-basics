```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1= pd.read_csv('Universities.csv')
```

```python
df1
```

```python
#To create a histogram
plt.figure(figsize=(4,4))
plt.hist(df1['Accept'])
plt.xlabel('Accept')
plt.ylabel('Count')
plt.title('Histogram of accept')
#to remove values & look only for graph
plt.show()
```

```python
#To create a histogram
plt.figure(figsize=(4,4))
plt.hist(df1['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('Histogram of SAT')
#to remove values & look only for graph
plt.show()
```

```python
#This is for both histogram & distribution
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df1['Accept'])
```

```python
plt.plot(df1['Accept'],df1['Expenses'],'r+')
plt.xlabel('Accept')
plt.ylabel('Expenses')
plt.title('Scatterplot vs accept')
```
